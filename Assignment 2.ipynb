```python
#6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'
```

```python
marks=82
```

```python
if marks>=90:
    print("Grade is A")
elif marks>=75 and marks<=89:
    print("Grade is B")
elif marks>=50 and marks<=74:
    print("Grade is C")
else:
    print("Fail")
```
