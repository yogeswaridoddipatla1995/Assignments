```python
#●	Apply scaling techniques to numerical features:
#a.	Standard Scaling   b. Min-Max Scaling
```

```python
df_count=df[df.columns[[0,2,4,10,11,12]]]
df_count
```

```python
#a.	Standard Scaling
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss_count=ss.fit_transform(df_count)
ss_count= pd.DataFrame(ss_count)
ss_count.columns= list(df_count)
ss_count
```

```python
#b. Min-Max Scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm_count=mm.fit_transform(df_count)
mm_count=pd.DataFrame(mm_count)
mm_count.columns=list(df_count)
mm_count
```

```python
#●	Discuss the scenarios where each scaling technique is preferred and why.
```

```python
# StandardScaler has mean 0 and standard deviation as 1 where as MinMaxScaler has the range (0,1) and it is also between 0 and 1.
# StandardScaler is used in maching learning model like logistic regression,linear regression,neural networks,pcm .
# MinMaxScaler is used in distance based models like DBSCAN,KNN,K-means.
# StandardScaler is used when data is following normal distribution.
```

```python
#●	Discuss the pros and cons of One-Hot Encoding and Label Encoding
```

```python
# We use One-Hot encoding if we have binary classification in our column.Here the output is in 0 and 1.
# We use Label Encoding if we have multi classification in our column.Here the output is o,1,2,3...... depending on our classifacation in our column.
# pros and cons for One-Hot encoding
# pros:
# improves performance of machine learning models.
# There will be no ranking like label encoding.
# cons:
# Due to One-Hot encoding number of columns increases.
# It slow down the process and memory consumption.
# pros and cons for Label Encoding
# pros
# produces a single column and less memory
# best for ordinal data.
# cons:
# Imposes an artificial order on nominal categories
# Models may incorrectly interpret encoded values as numerical magnitudes.
```

```python
# About capital_net:
# By creating capital_net features, we can get directly the total net. Here the captial_gain is how much he got profit when sells an asset or property 
# and capital_loss is how much he got loss in selling the property, now by getting capital_net we can now how much amount he is left with. 
# Which can be used in predicting his income.
```

```python
# About work_efficiency:
# We have age and hours per week columns seperately, by using these two we can get work_efficiency. 
# Here if the person is young he can do work more effieciently than a older man in less time.
# So work_efficieny varies according age which can be used in predicting his income.
```
