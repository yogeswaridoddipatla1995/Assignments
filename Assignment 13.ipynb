```python
#1.	Impute missing values.# There are no null values in original data set.
```

```python
df_new=df[df.columns[[0,1,2,3,4,5,6,7,8]]]
df_count=df[df.columns[[0,1,2,3,4,5,6,7]]]
```

```python
#2.	Encode categorical variables using one-hot encoding or label encoding.
# The categorical variable 'Outcome' is already encoded into one-hot encoding.
```

```python
#3.	If needed you can apply more preprocessing methods on the given dataset.
# But there are continous variables that should be standardized.
# Standardization of contionus variables.
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss_count= ss.fit_transform(df_count)
ss_count= pd.DataFrame(ss_count)
ss_count.columns=df_count.columns
```

```python
X= ss_count
Y= df['Outcome']
```

```python
# Building Predictive Models:
# 1.Split the preprocessed dataset into training and testing sets.
# 2.Choose appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score) for model evaluation.
# 3.Build predictive models using LightGBM and XGBoost algorithms.
# 4.Train the models on the training set and evaluate their performance on the testing set.
# 5.Use techniques like cross-validation and hyperparameter tuning to optimize model performance.
```

```python
# Building Predictive Models:
# 1.Split the preprocessed dataset into training and testing sets.
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
# 3.Build predictive models using LightGBM and XGBoost algorithms
# Predicting model using XGBoost algorithm
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=100,learning_rate=0.1,gamma=10,reg_lambda=4)
# 4.Train the models on the training set and evaluate their performance on the testing set.
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
# 2.Choose appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score) for model evaluation.
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(Y_train,Y_pred_train)
test_accuracy = accuracy_score(Y_test,Y_pred_test)
print(f"accuracy score for training data in XGB model is {np.round(train_accuracy,2)}")
print(f"accuracy score for testing data in XGB model is {np.round(test_accuracy,2)}")
```

```python
from sklearn.metrics import precision_score
train_precision = precision_score(Y_train,Y_pred_train)
test_precision = precision_score(Y_test,Y_pred_test)
print(f"precision_score for training data is {np.round(train_precision,2)}")
print(f"precision_score for testing data is {np.round(test_precision,2)}")
```

```python
from sklearn.metrics import recall_score
train_recall = recall_score(Y_train,Y_pred_train)
test_recall = recall_score(Y_test,Y_pred_test)
print(f"recall_score for training data is {np.round(train_recall,2)}")
print(f"recall_score for testing data is {np.round(test_recall,2)}")
```

```python
from sklearn.metrics import f1_score
train_f1 = f1_score(Y_train,Y_pred_train)
test_f1 = f1_score(Y_test,Y_pred_test)
print(f"f1_score for training data is {np.round(train_f1,2)}")
print(f"f1_score for testing data is {np.round(test_f1,2)}")
```

```python
# Hyper parameter tuning for lightgb model
from sklearn.model_selection import GridSearchCV
param_grid= ({
    'n_estimators': [100,150,200,],
    'learning_rate': [0.1,0.5,1.5],
    'gamma': [10,20,30],
    'reg_lambda': [4,6,8],
})
```

```python
grid_search = GridSearchCV(model,param_grid,scoring='accuracy',n_jobs=-1,verbose=2)
grid_search.fit(X_train,Y_train)
Y_pred_train = grid_search.predict(X_train)
Y_pred_test = grid_search.predict(X_test)
train_accuracy = accuracy_score(Y_train,Y_pred_train)
test_accuracy = accuracy_score(Y_test,Y_pred_test)
print(f"accuracy score for training data is {np.round(train_accuracy,2)}")
print(f"accuracy score for testing data is {np.round(test_accuracy,2)}")
```

```python
best_estimator = grid_search.best_estimator_
best_estimator
```

```python
# cross validation
from sklearn.model_selection import ShuffleSplit,cross_validate
model = XGBClassifier(n_estimators=100,learning_rate=0.1,gamma=10,reg_lambda=4)
shuffle_split= ShuffleSplit(n_splits=10,test_size=0.20)
cv_results = cross_validate(model,X,Y,cv=shuffle_split,scoring='accuracy',return_train_score=True)
train_scores = cv_results['train_score']
test_scores = cv_results['test_score']
print("cross validation: train accuracy is ",train_scores)
print("cross validation: test accuracy is ",test_scores)
```
